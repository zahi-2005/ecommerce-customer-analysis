```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_excel('../data/raw/online_retail_II.xlsx', sheet_name='Year 2009-2010')
print(df.shape)
print(df.head())
```

```python
print(df.isnull().sum())
```

```python
# Drop rows with missing Customer ID and Description
df = df.dropna(subset=['Customer ID', 'Description'])

# Remove cancelled orders (Invoice starts with 'C')
df = df[~df['Invoice'].astype(str).str.startswith('C')]

# Remove rows with negative or zero Quantity and Price
df = df[df['Quantity'] > 0]
df = df[df['Price'] > 0]

# Create Revenue column
df['Revenue'] = df['Quantity'] * df['Price']

# Convert InvoiceDate to datetime
df['InvoiceDate'] = pd.to_datetime(df['InvoiceDate'])

# Extract Year and Month
df['Year'] = df['InvoiceDate'].dt.year
df['Month'] = df['InvoiceDate'].dt.month
df['YearMonth'] = df['InvoiceDate'].dt.to_period('M')

print(df.shape)
print(df.head())
```

```python
df.to_csv('../data/cleaned/retail_cleaned.csv', index=False)
print("Saved successfully")
```
